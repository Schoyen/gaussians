```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gaussian(x, alpha=1, A_x=0, i=0):
    assert i >= 0

    if i == 0:
        return np.exp(-alpha * (x - A_x) ** 2)

    return (x - A_x) * gaussian(x, alpha=alpha, A_x=A_x, i=i - 1)
```

```python
def hermite_gaussian(x, alpha=1, A_x=0, i=0):

    if i < 0:
        return 0

    if i == 0:
        return np.exp(-alpha * (x - A_x) ** 2)

    return (
        2
        * alpha
        * (
            (x - A_x) * hermite_gaussian(x, alpha=alpha, A_x=A_x, i=i - 1)
            - (i - 1) * hermite_gaussian(x, alpha=alpha, A_x=A_x, i=i - 2)
        )
    )
```

```python
x = np.linspace(-3, 3, 101)

for i in range(5):
    plt.plot(x, gaussian(x, i=i), label=f"i={i}")

plt.grid()
plt.legend()
plt.show()
```

```python
x = np.linspace(-3, 3, 101)

for i in range(5):
    plt.plot(x, hermite_gaussian(x, i=i), label=f"i={i}")

plt.grid()
plt.legend()
plt.show()
```
